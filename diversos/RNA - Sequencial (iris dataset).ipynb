```python
#-*- encoding: utf-8 -*-

#Ajustar para que funcione com outras versões

import numpy as np
from sklearn import datasets
from sklearn.model_selection import train_test_split
from yellowbrick.classifier import ConfusionMatrix
from keras.models import Sequential
from keras.layers import Dense
from keras.utils import np_utils
from sklearn.metrics import confusion_matrix
```

```python
#Carregando a base de dados
base = datasets.load_iris()
previsores = base.data
classe = base.target
```

```python
#Ajuste para que a camada de saída tenha mais de um neurônio de acorco do as classes do problemas
classe_dummy = np_utils.to_categorical(classe)
```

```python
#Ajustando o dados de treinamento e validação/teste
x_treinamento, x_teste, y_treinamento, y_teste = train_test_split(previsores, classe_dummy, test_size = 0.3, random_state = 0)
```

```python
#Criando um modelo de RNA sequencial de 3 camadas. A função softmax se adequa mais ao problema
modelo = Sequential()
modelo.add(Dense(units = 5, input_dim = 4))
modelo.add(Dense(units = 4))
modelo.add(Dense(units = 3, activation='softmax'))
```

```python
#Vendo detalhes da RNA
modelo.summary()
```

```python
#Executando o modelo e treinando
modelo.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
modelo.fit(x_treinamento, y_treinamento, epochs=1000, validation_data = (x_teste, y_teste))
```

```python
#Fazendo as previsões eajustando a sua saída
previsoes = modelo.predict(x_teste)
previsoes = (previsoes > 0.5)
```

```python
#Vendo os resultados (lista comprehention)
y_teste_matrix = [np.argmax(t) for t in y_teste]
y_previsao_matrix = [np.argmax(t) for t in previsoes]
confusao = confusion_matrix(y_teste_matrix, y_previsao_matrix)
print(confusao)
```
