```python
# -*- encoding:utf-8 -*-

from igraph import Graph
from igraph import plot
```

```python
#Criando o grafo1
grafo1 = Graph(edges = [(0,1),(1,2),(2,3),(3,0)], directed = True)
grafo1.vs['label'] = range(grafo1.vcount())
plot(grafo1, bbox =(200,200))
```

```python
#Criando o grafo2
grafo2 = Graph(edges = [(0,1),(1,2),(2,3),(3,0),(0,3),(3,2),(2,1),(1,0),(1,3),(2,0),(3,1)], directed = True)
grafo2.vs['label'] = range(grafo2.vcount())
plot(grafo2, bbox =(200,200))
```

```python
#Criando o grafo3
grafo3 = Graph(edges = [(0,1),(1,1)], directed = True)
grafo3.add_vertex(5)
grafo3.vs['label'] = range(grafo3.vcount())
plot(grafo3, bbox =(200,200))
```
