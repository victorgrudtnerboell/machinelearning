```python
# -*- encoding:utf-8 -*-

#vs -> vertices
#es -> edges

from igraph import Graph
from igraph import plot
```

```python
grafo4 = Graph(directed=False)
grafo4.add_vertices(5)
grafo4.add_edges([(0,1),(1,2),(2,3),(3,4),(4,0),(0,2),(2,1)])
grafo4.vs['label'] = ['Abacate','Bolacha','Casa','Dado','Erro']
grafo4.vs['name'] = ['A','B','C','D','E']
print(grafo4.get_adjacency()[0,])
#adicionando peso
grafo4.vs['peso'] = [2,56,33,78,2]
for v in grafo4.vs:
    print(v)
print(grafo4.vs[2])
#Adicionando relacoes
grafo4.es['TipoAmizade'] = ['Amigo','Inimigo','Inimigo','Amigo','Jubarte']
grafo4.es['peso'] = [2,56,33,78,2]
print(grafo4.es[3])
print(grafo4)
#Nome do grafo
grafo4['nome'] = 'grafite'
plot(grafo4, bbox = (200,200))
```
