```python
# -*- encoding:utf-8 -*-

from igraph import Graph
from igraph import plot
```

```python
#Criando o grafo1
grafo1 = Graph(edges = [(0,1),(2,2),(2,3),(3,0)], directed = True)
grafo1.vs['label'] = range(grafo1.vcount())
plot(grafo1, bbox =(200,200))
```

```python
print(grafo1)
```

```python
#Criando o grafo1
grafo2 = Graph(edges = [(0,1),(2,2),(2,3),(3,0)], directed = False)
grafo2.vs['label'] = range(grafo2.vcount())
plot(grafo2, bbox =(200,200))
```

```python
print(grafo2)
```

```python
grafo3 = Graph(directed=False)
grafo3.add_vertices(10)
grafo3.add_vertex(16)
grafo3.add_edges([(0,1),(1,10),(1,5)])
plot(grafo3, bbox = (200,200))
```

```python
grafo4 = Graph(directed=False)
grafo4.add_vertices(5)
grafo4.add_edges([(0,1),(1,3),(1,4),(1,2)])
grafo4.vs['label'] = ['A','B','C','D','E'] 
plot(grafo4, bbox = (200,200))
```
