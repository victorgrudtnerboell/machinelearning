```python
# -*- encoding:utf-8 -*-

from igraph import Graph
from igraph import plot
import igraph #importar melhor
```

```python
#Carregando a grafo de um arquivo
grafo = igraph.load('dados/Grafo.graphml')
```

```python
plot(grafo, bbox=(300,300))
```

```python
#Grau do grfo
grafo.degree(type='all')
```

```python
#Grau de entrada do grfo
grafo.degree(type='in')
```

```python
#Grau de sada do grfo
grafo.degree(type='out')
```

```python
#Analisar influencia analisando o ggrau do vrtice
grau = grafo.degree(type='in')
plot(grafo, vertex_size = grau)
```

```python
#Diametro do grafo (medir a maior distncia entre os vrtices)
grafo.diameter(directed = False)
```

```python
#Diametro do grafo (medir so vrtices que tem maior distncia no grafo)
grafo.get_diameter()
```

```python
#Visualizar qual  a vizinhana
grafo.neighborhood()
```

```python
#Testando se os grfos so isomorfos
grafo2 = grafo
grafo.isomorphic(grafo2)
```
