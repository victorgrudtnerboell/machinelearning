```python
# -*- encoding:utf-8 -*-

from igraph import Graph
from igraph import plot
import igraph #importar melhor
```

```python
#Carregando a grafo de um arquivo
grafo = igraph.load('dados/Grafo.graphml')
```

```python
plot(grafo, bbox=(300,300))
```

```python
#Grau do gráfo
grafo.degree(type='all')
```

```python
#Grau de entrada do gráfo
grafo.degree(type='in')
```

```python
#Grau de saída do gráfo
grafo.degree(type='out')
```

```python
#Analisar influencia analisando o ggrau do vértice
grau = grafo.degree(type='in')
plot(grafo, vertex_size = grau)
```

```python
#Diametro do grafo (medir a maior distância entre os vértices)
grafo.diameter(directed = False)
```

```python
#Diametro do grafo (medir so vértices que tem maior distância no grafo)
grafo.get_diameter()
```

```python
#Visualizar qual é a vizinhança
grafo.neighborhood()
```

```python
#Testando se os gráfos são isomorfos
grafo2 = grafo
grafo.isomorphic(grafo2)
```
