```python
# -*- encoding:utf-8 -*-

from igraph import Graph
from igraph import plot
```

```python
grafo = Graph(edges = [(0,2),(0,1),(1,4),(1,5),(2,3),(6,7),(3,7),(4,7),(5,6)], directed= True)
grafo.vs['label'] = ['a','b','c','d','e','f','g','h','i']
grafo.es['weight'] = [2,1,2,1,2,1,3,1,3]
```

```python
plot(grafo, bbox = (300,300), edge_label = grafo.es['weight'])
```

```python
#Descobrindo a rota com menor distância (vpath = quais vertices vou visitar)
caminho_vertice = grafo.get_shortest_paths(0,7,output='vpath')
```

```python
print(caminho_vertice)
```

```python
for n in caminho_vertice[0]:
    print(grafo.vs[n]['label'])
```

```python
#Descobrindo a rota com menor distância (vpath = quais arestas (edges) vou visitar)
caminho_aresta = grafo.get_shortest_paths(0,7,output='epath')
print(caminho_aresta)
for n in caminho_aresta[0]:
    print(grafo.es[n]['weight'])
```

```python
#Distancia
caminho_aresta_id = []
for n in caminho_aresta[0]:
    caminho_aresta_id.append(n)
distancia = 0
for e in grafo.es:
    if e.index in caminho_aresta_id:
        distancia += grafo.es[e.index]['weight']
```

```python
distancia
```

```python
#Mostrando o caminho
caminho_nome_vertices = []
for n in caminho_vertice[0]:
    print(grafo.vs['label'])
    caminho_nome_vertices.append(grafo.vs[n]['label'])
#Colorindo vertices
for v in grafo.vs:
    if v['label'] in caminho_nome_vertices:
        v['color'] = 'green'
    else:
        v['color'] = 'gray'
        
#Colorindo as arestas
for e in grafo.es:
    if e.index in caminho_aresta_id:
        e['color'] = 'green'
    else:
        e['color'] = 'gray'
plot(grafo, bbox = (300,300), edge_label = grafo.es['weight'])
```
