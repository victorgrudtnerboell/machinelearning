```python
# -*- encoding:utf-8 -*-

from igraph import Graph
from igraph import plot
import igraph #importar melhor
import numpy as np
```

```python
#Carregando a grafo de um arquivo
grafo = igraph.load('dados/Grafo.graphml')
```

```python
#Verificando comunidades (cluster)
comunidades = grafo.clusters()
#print(comunidades)
#print(comunidades.membership)
plot(grafo, vertex_color = comunidades.membership)
```

```python
#Descobrindo comunidades
comunidades2 = grafo.clusters()
#print(comunidades2)
c = grafo.community_edge_betweenness()
#print(c)
#print(c.optimal_count)
comunidades3 = c.as_clustering()
#print(comunidades3)
print(comunidades3.membership)
cores = comunidades3.membership
cores = np.array(cores)
cores = cores*20
cores = cores.tolist()
print(cores)
```

```python
plot(grafo, bbox=(500,500), vertex_color = cores)
```

```python
#Vendo o número de cliques
cli = grafo.as_undirected().cliques(min=4)
print(cli)
```
